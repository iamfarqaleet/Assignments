### Statistics Advance 1 Theoratical Questions

Q1. What is a random variable in probability theory?
- A random variable is a numerical outcome of a random phenomenon. It is a function that assigns a real number to each outcome in a sample space, allowing us to quantify the results of random experiments.

Q2. What are the types of random variables?
- There are two main types of random variables:
  1. Discrete Random Variables: These can take on a countable number of distinct values (e.g., the number of heads in a series of coin flips).
  2. Continuous Random Variables: These can take on any value within a given range (e.g., the height of individuals in a population).
  

Q3. What is the difference between discrete and continuous distributions?
- Discrete distributions describe the probabilities of discrete random variables, which can take on specific, countable values (e.g., the number of students in a class). Continuous distributions describe the probabilities of continuous random variables, which can take on any value within a range (e.g., the weight of individuals).


Q4. What are probability distribution functions (PDF)?
- Probability Distribution Functions (PDF) are mathematical functions that describe the likelihood of different outcomes for a random variable. For discrete random variables, the PDF gives the probability of each possible value. For continuous random variables, the PDF describes the density of probabilities over a range of values, and the area under the curve represents probabilities.


Q5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
- Cumulative Distribution Functions (CDF) provide the cumulative probability that a random variable takes on a value less than or equal to a specific point. In contrast, Probability Distribution Functions (PDF) give the probability of the random variable taking on a specific value. The CDF is the integral of the PDF for continuous variables and the sum for discrete variables.


Q6. What is a discrete uniform distribution?
- A discrete uniform distribution is a probability distribution where all possible outcomes are equally likely. In this distribution, each value has the same probability of occurring, and it is often used when dealing with a finite set of outcomes (e.g., rolling a fair die).


Q7. What are the key properties of a Bernoulli distribution?
- The key properties of a Bernoulli distribution include:
  1. It has only two possible outcomes: success (1) and failure (0).
  2. The probability of success is denoted by \( p \), and the probability of failure is \( 1 - p \).
  3. The mean of a Bernoulli distribution is \( p \), and the variance is \( p(1 - p) \).


Q8. What is the binomial distribution, and how is it used in probability?
- The binomial distribution is a discrete probability distribution that describes the number of successes in a fixed number of independent Bernoulli trials, each with the same probability of success \( p \). It is used to model scenarios where there are a fixed number of trials, each with two possible outcomes (success or failure). The probability mass function (PMF) for the binomial distribution is given by:
\[P(X = k) = \binom{n}{k} p^k (1 - p)^{n-k}\]
where \( n \) is the number of trials, \( k \) is the number of successes, and \( \binom{n}{k} \) is the binomial coefficient.


Q9. What is the Poisson distribution and where is it applied?
- The Poisson distribution is a discrete probability distribution that describes the number of events occurring in a fixed interval of time or space, given that these events happen with a known constant mean rate and independently of the time since the last event. It is often used in scenarios such as modeling the number of phone calls received at a call center in an hour or the number of decay events from a radioactive source in a given time period. The probability mass function (PMF) for the Poisson distribution is given by:
\[P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}\]
where \( \lambda \) is the average rate of occurrence, \( k \) is the number of events, and \( e \) is Euler's number (approximately 2.71828).


Q10. What is a continuous uniform distribution?
- A continuous uniform distribution is a probability distribution where all outcomes in a given interval are equally likely. In this distribution, the probability density function (PDF) is constant over the interval, and the total area under the PDF curve equals 1. It is often used to model scenarios where any value within a specified range is equally probable (e.g., selecting a random number between 0 and 1). The PDF for a continuous uniform distribution on the interval \([a, b]\) is given by:
\[f(x) = \frac{1}{b - a}\]


Q11. What are the characteristics of a normal distribution?
- The characteristics of a normal distribution include:
  1. It is symmetric about its mean, which is also the median and mode.
  2. The mean, median, and mode are all equal.
  3. The total area under the curve equals 1.
  4. Approximately 68% of the data falls within one standard deviation from the mean, about 95% falls within two standard deviations, and about 99.7% falls within three standard deviations (the empirical rule).
  5. It has a bell-shaped curve, with tails that approach but never touch the horizontal axis.
  

Q12. What is the standard normal distribution, and why is it important?
- The standard normal distribution is a special case of the normal distribution with a mean of 0 and a standard deviation of 1. It is important because it serves as a reference distribution for all normal distributions. Any normal distribution can be transformed into a standard normal distribution using the z-score formula:
\[z = \frac{X - \mu}{\sigma}\]
where \( X \) is the value, \( \mu \) is the mean, and \( \sigma \) is the standard deviation. This transformation allows for easier comparison and calculation of probabilities across different normal distributions.


Q13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
- The Central Limit Theorem (CLT) states that the distribution of the sample means approaches a normal distribution as the sample size increases, regardless of the shape of the population distribution, provided that the samples are independent and identically distributed. This is critical in statistics because it allows us to make inferences about population parameters using sample statistics, enabling hypothesis testing and confidence interval estimation even when the underlying population distribution is not normal.


Q14. How does the Central Limit Theorem relate to the normal distribution?
- The Central Limit Theorem (CLT) relates to the normal distribution by stating that as the sample size increases, the distribution of the sample means will approximate a normal distribution, regardless of the original population's distribution. This means that even if the population is not normally distributed, the sampling distribution of the mean will become normal as long as the sample size is sufficiently large (typically \( n \geq 30 \)). This property allows statisticians to apply normal distribution techniques to infer about population parameters based on sample data.


Q15. What is the application of Z statistics in hypothesis testing?
- Z statistics are used in hypothesis testing to determine whether to reject or fail to reject a null hypothesis based on the standard normal distribution. When the population standard deviation is known, Z statistics can be calculated using the formula:
\[Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}}\]
where \( \bar{X} \) is the sample mean, \( \mu \) is the population mean under the null hypothesis, \( \sigma \) is the population standard deviation, and \( n \) is the sample size. The Z statistic is then compared to critical values from the standard normal distribution to make decisions about the null hypothesis.


Q16. How do you calculate a Z-score, and what does it represent?
- A Z-score is calculated using the formula:
\[Z = \frac{X - \mu}{\sigma}\]
where \( X \) is the value, \( \mu \) is the mean of the population, and \( \sigma \) is the standard deviation of the population. The Z-score represents the number of standard deviations a data point is from the mean. A positive Z-score indicates that the value is above the mean, while a negative Z-score indicates that it is below the mean.


Q17. What are point estimates and interval estimates in statistics?
- Point estimates are single values that serve as the best guess or estimate of a population parameter (e.g., the sample mean as an estimate of the population mean). Interval estimates, on the other hand, provide a range of values within which the population parameter is expected to lie, along with a specified level of confidence (e.g., a 95% confidence interval for the population mean). Interval estimates account for sampling variability and provide more information about the uncertainty associated with the estimate.


Q18. What is the significance of confidence intervals in statistical analysis?
- Confidence intervals are significant in statistical analysis because they provide a range of values within which the true population parameter is likely to fall, with a specified level of confidence (e.g., 95% or 99%). They help quantify the uncertainty associated with point estimates and allow researchers to make informed decisions based on sample data. Confidence intervals also facilitate comparisons between different groups or treatments by indicating the precision of the estimates.


Q19. What is the relationship between a Z-score and a confidence interval?
- The relationship between a Z-score and a confidence interval lies in the fact that Z-scores are used to determine the critical values that define the boundaries of the confidence interval. For a given confidence level (e.g., 95% or 99%), the Z-score corresponding to that level is used to calculate the margin of error, which is then added to and subtracted from the point estimate (e.g., sample mean) to create the confidence interval. The formula for a confidence interval is:
\[ \text{Confidence Interval} = \bar{X} \pm Z \left(\frac{\sigma}{\sqrt{n}}\right) \]
where \( \bar{X} \) is the sample mean, \( Z \) is the Z-score for the desired confidence level, \( \sigma \) is the population standard deviation, and \( n \) is the sample size. This interval provides a range of values that likely contains the true population parameter with a specified level of confidence.


Q20. How are Z-scores used to compare different distributions?
- Z-scores are used to compare different distributions by standardizing the values from different datasets to a common scale. By converting raw scores into Z-scores, we can determine how many standard deviations a value is from the mean of its distribution. This allows for meaningful comparisons between values from different distributions, even if they have different means and standard deviations. For example, a Z-score of 2 indicates that a value is 2 standard deviations above the mean, regardless of the original distribution's parameters.


Q21. What are the assumptions for applying the Central Limit Theorem?
- The assumptions for applying the Central Limit Theorem (CLT) include:
  1. **Independence**: The samples must be independent of each other.
  2. **Identically Distributed**: The samples should be drawn from the same population or have the same distribution.
  3. **Sample Size**: The sample size should be sufficiently large (typically \( n \geq 30 \)) for the CLT to hold, especially if the population distribution is not normal.
  4. **Finite Variance**: The population from which the samples are drawn should have a finite variance.
  

Q22. What is the concept of expected value in a probability distribution?
- The expected value (or mean) of a probability distribution is a measure of the central tendency of the distribution. It represents the long-term average outcome of a random variable if the experiment were repeated many times. For a discrete random variable, the expected value is calculated as:
\[E(X) = \sum_{i=1}^{n} x_i P(x_i)\]
where \( x_i \) are the possible values of the random variable and \( P(x_i) \) is the probability of each value. For a continuous random variable, the expected value is calculated using the integral of the product of the value and its probability density function (PDF):
\[E(X) = \int_{-\infty}^{\infty} x f(x) dx\]
where \( f(x) \) is the PDF of the continuous random variable.


Q23. How does a probability distribution relate to the expected outcome of a random variable?
- A probability distribution describes the likelihood of different outcomes for a random variable, and the expected outcome (or expected value) is a summary measure that represents the average value of the random variable over many trials. The expected value is calculated using the probabilities assigned to each possible outcome in the distribution. Thus, the expected value provides a single number that captures the central tendency of the distribution, while the probability distribution provides a complete picture of all possible outcomes and their associated probabilities.

```python
# Q2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).

import matplotlib.pyplot as plt
import seaborn as sns


def plot_discrete_uniform_distribution(a, b):
    # Generate a discrete uniform distribution
    x = np.arange(a, b + 1)
    pmf = np.ones_like(x) / len(x)  # PMF for discrete uniform distribution

    # Plot the PMF
    plt.figure(figsize=(8, 5))
    sns.barplot(x=x, y=pmf, color="blue", alpha=0.7)
    plt.title("Probability Mass Function of Discrete Uniform Distribution")
    plt.xlabel("Random Variable")
    plt.ylabel("Probability")
    plt.xticks(x)
    plt.ylim(0, 0.2)
    plt.grid(axis="y")
    plt.show()


# Example usage
plot_discrete_uniform_distribution(1, 10)
```

```python
# Q3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.


def bernoulli_pdf(p, x):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    p (float): Probability of success (0 <= p <= 1)
    x (int): Random variable (0 or 1)

    Returns:
    float: Probability of the random variable
    """
    if x not in [0, 1]:
        raise ValueError("x must be either 0 or 1 for Bernoulli distribution.")

    return p if x == 1 else (1 - p)


# Example usage
p = 0.7  # Probability of success
x_values = [0, 1]
pdf_values = [bernoulli_pdf(p, x) for x in x_values]
# Print the PDF values
for x, pdf in zip(x_values, pdf_values):
    print(f"PDF of Bernoulli distribution for x={x}: {pdf:.2f}")
```

```python
# Q6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.


def plot_discrete_uniform_cdf(a, b):
    """
    Calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.

    Parameters:
    a (int): Lower bound of the distribution
    b (int): Upper bound of the distribution

    Returns:
    None
    """
    x = np.arange(a, b + 1)
    cdf = np.cumsum(np.ones_like(x) / len(x))  # CDF for discrete uniform distribution

    # Plot the CDF
    plt.figure(figsize=(8, 5))
    sns.lineplot(x=x, y=cdf, marker="o", color="blue", alpha=0.7)
    plt.title("Cumulative Distribution Function of Discrete Uniform Distribution")
    plt.xlabel("Random Variable")
    plt.ylabel("Cumulative Probability")
    plt.xticks(x)
    plt.grid()
    plt.show()


# Example usage
plot_discrete_uniform_cdf(1, 10)
```

```python
# Q9. Write a Python function to calculate Z-scores from a dataset and plot them.


def calculate_z_scores(data):
    """
    Calculate Z-scores for a given dataset.

    Parameters:
    data (array-like): Input data

    Returns:
    array: Z-scores of the input data
    """
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = (data - mean) / std_dev
    return z_scores


def plot_z_scores(data):
    """Plot Z-scores of a given dataset.
    Parameters:
    data (array-like): Input data
    Returns:
    None
    """
    z_scores = calculate_z_scores(data)

    # Plot the Z-scores
    plt.figure(figsize=(8, 5))
    sns.histplot(
        z_scores,
        bins=30,
        kde=True,
        stat="density",
        color="blue",
        alpha=0.7,
    )
    plt.title("Z-Scores of the Dataset")
    plt.xlabel("Z-Score")
    plt.ylabel("Density")
    plt.grid(axis="y")
    plt.show()


# Example usage
data = np.random.normal(0, 1, 1000)  # Generate random data from a normal distribution
plot_z_scores(data)
```

```python
# Q10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.


def central_limit_theorem(samples, num_samples=1000, sample_size=30):
    """
    Implement the Central Limit Theorem for a non-normal distribution.

    Parameters:
    samples (array-like): Input samples from a non-normal distribution
    num_samples (int): Number of samples to draw
    sample_size (int): Size of each sample

    Returns:
    array: Sample means
    """
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.choice(samples, size=sample_size, replace=True)
        sample_means.append(np.mean(sample))

    return np.array(sample_means)


def plot_clt(samples, num_samples=1000, sample_size=30):
    """Plot the results of the Central Limit Theorem.
    Parameters:
    samples (array-like): Input samples from a non-normal distribution
    num_samples (int): Number of samples to draw
    sample_size (int): Size of each sample
    Returns:
    None
    """
    sample_means = central_limit_theorem(samples, num_samples, sample_size)

    # Plot the histogram of sample means
    plt.figure(figsize=(8, 5))
    sns.histplot(
        sample_means,
        bins=30,
        kde=True,
        stat="density",
        color="blue",
        alpha=0.7,
    )
    plt.title("Central Limit Theorem: Distribution of Sample Means")
    plt.xlabel("Sample Mean")
    plt.ylabel("Density")
    plt.grid(axis="y")
    plt.show()


# Example usage
non_normal_samples = np.random.exponential(
    scale=1, size=1000
)  # Generate non-normal samples from an exponential distribution
plot_clt(non_normal_samples, num_samples=1000, sample_size=30)
```

```python
# Q13. Generate random variables and calculate their corresponding probabilities using the binomial distribution.


def binomial_distribution(n, p, num_samples=1000):
    """
    Generate random variables and calculate their corresponding probabilities using the binomial distribution.

    Parameters:
    n (int): Number of trials
    p (float): Probability of success
    num_samples (int): Number of samples to generate

    Returns:
    tuple: Random variables and their corresponding probabilities
    """
    samples = np.random.binomial(n, p, num_samples)
    probabilities = np.bincount(samples) / num_samples  # Calculate probabilities

    return samples, probabilities


def plot_binomial_distribution(n, p, num_samples=1000):
    """Plot the binomial distribution with given parameters.
    Parameters:
    n (int): Number of trials
    p (float): Probability of success
    num_samples (int): Number of samples to generate
    Returns:
    None
    """
    samples, probabilities = binomial_distribution(n, p, num_samples)

    # Plot the histogram
    plt.figure(figsize=(8, 5))
    sns.barplot(
        x=np.arange(len(probabilities)),
        y=probabilities,
        color="blue",
        alpha=0.7,
    )
    plt.title(f"Binomial Distribution (n={n}, p={p})")
    plt.xlabel("Number of Successes")
    plt.ylabel("Probability")
    plt.xticks(np.arange(len(probabilities)))
    plt.grid(axis="y")
    plt.show()


# Example usage
n = 10  # Number of trials
p = 0.5  # Probability of success
plot_binomial_distribution(n, p, num_samples=1000)
```

```python
# Q14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.


def calculate_z_score(data_point, mean, std_dev):
    """
    Calculate the Z-score for a given data point.

    Parameters:
    data_point (float): The data point for which to calculate the Z-score
    mean (float): Mean of the distribution
    std_dev (float): Standard deviation of the distribution

    Returns:
    float: Z-score of the data point
    """
    return (data_point - mean) / std_dev


def compare_z_score_with_standard_normal(data_point, mean=0, std_dev=1):
    """
    Compare the Z-score of a data point with the standard normal distribution.
    Parameters:
    data_point (float): The data point for which to calculate the Z-score
    mean (float): Mean of the standard normal distribution (default is 0)
    std_dev (float): Standard deviation of the standard normal distribution (default is 1)
    Returns:
    None
    """
    z_score = calculate_z_score(data_point, mean, std_dev)

    # Plot the standard normal distribution
    x = np.linspace(-4, 4, 1000)
    pdf = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(
        -0.5 * ((x - mean) / std_dev) ** 2
    )

    plt.figure(figsize=(8, 5))
    plt.plot(x, pdf, color="blue", label="Standard Normal Distribution")
    plt.axvline(z_score, color="red", linestyle="--", label=f"Z-score: {z_score:.2f}")
    plt.title("Comparison of Z-score with Standard Normal Distribution")
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid()
    plt.show()


# Example usage
data_point = 1.5  # Example data point
mean = 0  # Mean of the standard normal distribution
std_dev = 1  # Standard deviation of the standard normal distribution
compare_z_score_with_standard_normal(data_point, mean, std_dev)
```

```python
# Q15. Implement hypothesis testing using Z-statistics for a sample dataset.


def z_test(sample_mean, population_mean, std_dev, sample_size):
    """
    Perform a Z-test for a sample dataset.

    Parameters:
    sample_mean (float): Mean of the sample
    population_mean (float): Mean of the population
    std_dev (float): Standard deviation of the population
    sample_size (int): Size of the sample

    Returns:
    float: Z-statistic
    """
    z_statistic = (sample_mean - population_mean) / (std_dev / np.sqrt(sample_size))
    return z_statistic


def perform_z_test(sample_mean, population_mean, std_dev, sample_size, alpha=0.05):
    """
    Perform a Z-test and interpret the result.
    Parameters:
    sample_mean (float): Mean of the sample
    population_mean (float): Mean of the population
    std_dev (float): Standard deviation of the population
    sample_size (int): Size of the sample
    alpha (float): Significance level (default is 0.05)
    Returns:
    None
    """
    z_statistic = z_test(sample_mean, population_mean, std_dev, sample_size)
    critical_value = 1.96  # For a two-tailed test at alpha=0.05

    print(f"Z-statistic: {z_statistic:.2f}")
    if abs(z_statistic) > critical_value:
        print("Reject the null hypothesis.")
    else:
        print("Fail to reject the null hypothesis.")


# Example usage
sample_mean = 5.5  # Mean of the sample
population_mean = 5.0  # Mean of the population
std_dev = 1.0  # Standard deviation of the population
sample_size = 30  # Size of the sample
perform_z_test(sample_mean, population_mean, std_dev, sample_size, alpha=0.05)
```

```python
# Q16. Create a confidence interval for a dataset using Python and interpret the result.


def confidence_interval(sample_mean, std_dev, sample_size, confidence_level=0.95):
    """
    Calculate the confidence interval for a sample dataset.

    Parameters:
    sample_mean (float): Mean of the sample
    std_dev (float): Standard deviation of the sample
    sample_size (int): Size of the sample
    confidence_level (float): Confidence level (default is 0.95)

    Returns:
    tuple: Lower and upper bounds of the confidence interval
    """
    z_score = 1.96  # Z-score for 95% confidence level
    margin_of_error = z_score * (std_dev / np.sqrt(sample_size))

    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound


def plot_confidence_interval(sample_mean, std_dev, sample_size, confidence_level=0.95):
    """
    Plot the confidence interval for a sample dataset.
    Parameters:
    sample_mean (float): Mean of the sample
    std_dev (float): Standard deviation of the sample
    sample_size (int): Size of the sample
    confidence_level (float): Confidence level (default is 0.95)
    Returns:
    None
    """
    lower_bound, upper_bound = confidence_interval(
        sample_mean, std_dev, sample_size, confidence_level
    )

    # Plot the confidence interval
    plt.figure(figsize=(8, 5))
    plt.axhline(sample_mean, color="blue", linestyle="--", label="Sample Mean")
    plt.fill_betweenx(
        [0, 1],
        lower_bound,
        upper_bound,
        color="lightblue",
        alpha=0.5,
        label=f"{confidence_level*100:.1f}% Confidence Interval",
    )
    plt.title("Confidence Interval for Sample Mean")
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.legend()
    plt.grid()
    plt.show()


# Example usage
sample_mean = 5.5  # Mean of the sample
std_dev = 1.0  # Standard deviation of the sample
sample_size = 30  # Size of the sample
confidence_level = 0.95  # Confidence level
lower_bound, upper_bound = confidence_interval(
    sample_mean, std_dev, sample_size, confidence_level
)
print(f"Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
plot_confidence_interval(sample_mean, std_dev, sample_size, confidence_level)
```

```python
# Q17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.


def generate_normal_data_and_confidence_interval(
    mean, std_dev, sample_size, confidence_level=0.95
):
    """
    Generate data from a normal distribution and calculate the confidence interval for its mean.

    Parameters:
    mean (float): Mean of the normal distribution
    std_dev (float): Standard deviation of the normal distribution
    sample_size (int): Size of the sample
    confidence_level (float): Confidence level (default is 0.95)

    Returns:
    tuple: Sample mean, lower bound, and upper bound of the confidence interval
    """
    samples = np.random.normal(mean, std_dev, sample_size)
    sample_mean = np.mean(samples)
    lower_bound, upper_bound = confidence_interval(
        sample_mean, std_dev, sample_size, confidence_level
    )

    return sample_mean, lower_bound, upper_bound


def plot_normal_data_and_confidence_interval(
    mean, std_dev, sample_size, confidence_level=0.95
):
    """
    Plot the normal data and its confidence interval.

    Parameters:
    mean (float): Mean of the normal distribution
    std_dev (float): Standard deviation of the normal distribution
    sample_size (int): Size of the sample
    confidence_level (float): Confidence level (default is 0.95)

    Returns:
    None
    """
    sample_mean, lower_bound, upper_bound = (
        generate_normal_data_and_confidence_interval(
            mean, std_dev, sample_size, confidence_level
        )
    )

    # Plot the histogram of the samples
    samples = np.random.normal(mean, std_dev, sample_size)
    plt.figure(figsize=(8, 5))
    sns.histplot(
        samples,
        bins=30,
        kde=True,
        stat="density",
        color="blue",
        alpha=0.7,
    )

    # Plot the confidence interval
    plt.axvline(sample_mean, color="red", linestyle="--", label="Sample Mean")
    plt.fill_betweenx(
        [0, 1],
        lower_bound,
        upper_bound,
        color="lightblue",
        alpha=0.5,
        label=f"{confidence_level*100:.1f}% Confidence Interval",
    )

    plt.title("Normal Distribution with Confidence Interval")
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.legend()
    plt.grid()
    plt.show()


# Example usage
mean = 5.0  # Mean of the normal distribution
std_dev = 1.0  # Standard deviation of the normal distribution
sample_size = 30  # Size of the sample
confidence_level = 0.95  # Confidence level
sample_mean, lower_bound, upper_bound = generate_normal_data_and_confidence_interval(
    mean, std_dev, sample_size, confidence_level
)
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
plot_normal_data_and_confidence_interval(mean, std_dev, sample_size, confidence_level)
```

```python
# Q18. Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution.


def plot_normal_pdf(mean, std_dev, num_samples=1000):
    """
    Calculate and visualize the probability density function (PDF) of a normal distribution.

    Parameters:
    mean (float): Mean of the normal distribution
    std_dev (float): Standard deviation of the normal distribution
    num_samples (int): Number of samples to generate for plotting

    Returns:
    None
    """
    x = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, num_samples)
    pdf = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(
        -0.5 * ((x - mean) / std_dev) ** 2
    )

    # Plot the PDF
    plt.figure(figsize=(8, 5))
    plt.plot(x, pdf, color="blue", label="Normal Distribution PDF")
    plt.title(
        f"Probability Density Function of Normal Distribution (={mean}, ={std_dev})"
    )
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.legend()
    plt.grid()
    plt.show()


# Example usage
mean = 0  # Mean of the normal distribution
std_dev = 1  # Standard deviation of the normal distribution
plot_normal_pdf(mean, std_dev, num_samples=1000)
```

```python
# Q20. Simulate a random variable using a continuous uniform distribution and calculate its expected value.


def simulate_continuous_uniform(a, b, num_samples=1000):
    """
    Simulate a random variable using a continuous uniform distribution and calculate its expected value.

    Parameters:
    a (float): Lower bound of the distribution
    b (float): Upper bound of the distribution
    num_samples (int): Number of samples to generate

    Returns:
    tuple: Simulated samples and their expected value
    """
    samples = np.random.uniform(a, b, num_samples)
    expected_value = (a + b) / 2  # Expected value for continuous uniform distribution
    return samples, expected_value


def plot_continuous_uniform_with_expected_value(a, b, num_samples=1000):
    """
    Plot the continuous uniform distribution and its expected value.
    Parameters:
    a (float): Lower bound of the distribution
    b (float): Upper bound of the distribution
    num_samples (int): Number of samples to generate
    Returns:
    None
    """
    samples, expected_value = simulate_continuous_uniform(a, b, num_samples)

    # Plot the histogram of the samples
    plt.figure(figsize=(8, 5))
    sns.histplot(
        samples,
        bins=30,
        kde=True,
        stat="density",
        color="blue",
        alpha=0.7,
    )

    # Plot the expected value
    plt.axvline(
        expected_value,
        color="red",
        linestyle="--",
        label=f"Expected Value: {expected_value:.2f}",
    )
    plt.title(f"Continuous Uniform Distribution ({a}, {b}) with Expected Value")
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(axis="y")
    plt.show()


# Example usage
a = 0  # Lower bound of the distribution
b = 10  # Upper bound of the distribution
plot_continuous_uniform_with_expected_value(a, b, num_samples=1000)
```

```python
# Q22. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.


def calculate_range_and_iqr(data):
    """
    Calculate the range and interquartile range (IQR) of a dataset.

    Parameters:
    data (array-like): Input dataset

    Returns:
    tuple: Range and IQR of the dataset
    """
    data_range = np.max(data) - np.min(data)
    q75, q25 = np.percentile(data, [75, 25])
    iqr = q75 - q25
    return data_range, iqr


def plot_range_and_iqr(data):
    """
    Plot the range and interquartile range (IQR) of a dataset.
    Parameters:
    data (array-like): Input dataset
    Returns:
    None
    """
    data_range, iqr = calculate_range_and_iqr(data)

    # Plot the boxplot to visualize the IQR
    plt.figure(figsize=(8, 5))
    sns.boxplot(data=data, color="lightblue")
    plt.title("Boxplot of Dataset with Range and IQR")
    plt.xlabel("Dataset")
    plt.ylabel("Value")

    # Display range and IQR
    plt.text(0.05, 0.95, f"Range: {data_range:.2f}", transform=plt.gca().transAxes)
    plt.text(0.05, 0.90, f"IQR: {iqr:.2f}", transform=plt.gca().transAxes)

    plt.grid(axis="y")
    plt.show()


# Example usage
data = np.random.normal(0, 1, 1000)  # Generate random data from a normal distribution
data_range, iqr = calculate_range_and_iqr(data)
print(f"Range: {data_range:.2f}")
print(f"IQR: {iqr:.2f}")
plot_range_and_iqr(data)
```

```python
# Q23. Implement Z-score normalization on a dataset and visualize its transformation.


def z_score_normalization(data):
    """
    Apply Z-score normalization to a dataset.

    Parameters:
    data (array-like): Input dataset

    Returns:
    array: Z-score normalized dataset
    """
    mean = np.mean(data)
    std_dev = np.std(data)
    normalized_data = (data - mean) / std_dev
    return normalized_data


def plot_z_score_normalization(data):
    """
    Plot the original and Z-score normalized datasets.
    Parameters:
    data (array-like): Input dataset
    Returns:
    None
    """
    normalized_data = z_score_normalization(data)

    # Plot the original and normalized datasets
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    sns.histplot(data, bins=30, kde=True, color="blue", alpha=0.7)
    plt.title("Original Dataset")
    plt.xlabel("Value")
    plt.ylabel("Density")

    plt.subplot(1, 2, 2)
    sns.histplot(normalized_data, bins=30, kde=True, color="orange", alpha=0.7)
    plt.title("Z-score Normalized Dataset")
    plt.xlabel("Z-score")
    plt.ylabel("Density")

    plt.tight_layout()
    plt.show()


# Example usage
data = np.random.normal(0, 1, 1000)  # Generate random data from a normal distribution
normalized_data = z_score_normalization(data)
print(f"First 10 Z-score Normalized Values: {normalized_data[:10]}")
plot_z_score_normalization(data)
```

```python
# Q24. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

import numpy as np
from scipy.stats import skew, kurtosis


def calculate_skewness_and_kurtosis():
    """
    Generate data from a normal distribution and calculate its skewness and kurtosis.

    Returns:
    tuple: Skewness and kurtosis of the generated dataset
    """
    # Generate random data from a normal distribution (mean=0, std=1)
    data = np.random.normal(loc=0, scale=1, size=1000)

    # Calculate skewness and kurtosis
    data_skewness = skew(data)
    data_kurtosis = kurtosis(data)  # Default is excess kurtosis (normal = 0)

    return data_skewness, data_kurtosis


# Example usage
skewness, kurt = calculate_skewness_and_kurtosis()
print(f"Skewness: {skewness:.4f}")
print(f"Kurtosis: {kurt:.4f}")
```
